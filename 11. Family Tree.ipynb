```python
family = {
    "John": {"Megan": "wife", "William": "son", "David": "son", "Adam": "son"},
    "William": {"Emma": "wife", "Chris": "son", "Stepanie": "son"},
    "David": {"Olivia": "wife", "Wayne": "son", "Tiffany": "son", "Julie": "daughter", "Neil": "son", "Peter": "son"},
    "Adam": {"Lily": "wife", "Sophia": "daughter"}
}


# print(directRelation("Adam","John"))

def relation(person1, person2):
    for person3 in family:
        if person1 in family[person3] and person2 in family[person3]:
            if family[person3][person1]=="son":
                if family[person3][person2]=="son":
                    return "brother"
                if family[person3][person2]=="daughter":
                    return "brother"
                if family[person3][person2]=="wife":
                    return "son"
            elif family[person3][person1]=="daughter":
                if family[person3][person2]=="son":
                    return "sister"
                if family[person3][person2]=="daughter":
                    return "sister"
                if family[person3][person2]=="wife":
                    return "daughter"
            elif family[person3][person1]=="wife":
                if family[person3][person2]=="son":
                    return "son"
                if family[person3][person2]=="daughter":
                    return "daughter"
                if family[person3][person2]=="wife":
                    return "sotin"
    # Direct Relation
    if person2 in family:
        if person1 in family[person2]:
            return family[person2][person1]
        else:
            for person3 in family:
                if person1 != person3 and person2 != person3:
                    if person1 in family[person3] and person3 in family[person2]:
                        if family[person2][person3]=="son":
                            if family[person3][person1]=="son":
                                return "grand son"
                            elif family[person3][person1]=="daughter":
                                return "grand daughter"
                            elif family[person3][person1]=="wife":
                                return "daughter in law"
    if person1 in family:
        if relation(person2, person1)=="son" or relation(person2, person1)=="daughter":
            return "father"
        elif relation(person2, person1)=="son in law" or relation(person2, person1)=="daughter in law":
            return "father in law"
        elif relation(person2, person1)=="wife":
            return "husband"
        elif relation(person2, person1)=="grand son" or relation(person2, person1)=="grand daughter":
            return "grand father"
        elif relation(person2, person1)=="daughter":
            return "father"
    for person3 in family:
        if relation(person2, person3)=="son":
            if relation(person1, person3)=="grand son":
                return "uncle"
            elif relation(person1, person3)=="daughter in law":
                return "sister in law"
    return  "none"

print("John", "William", relation("John", "William"))
print("William", "John", relation("William", "John"))
print("David", "Adam", relation("David", "Adam"))
print("Neil", "John", relation("Neil", "John"))
print("John", "Neil", relation("John", "Neil"))
print("Lily", "John",relation("Lily", "John"))
print("John", "Lily", relation("John", "Lily"))
print("Neil", "William", relation("Neil", "William"))
print("William", "Peter", relation("William", "Peter"))
print("Lily", "William", relation("Lily", "William"))

# f=input("Enter person1 person: ")
# l=input("Enter person2 person: ")
# print(relation(f,l))

# print(relation("John","Emma"))
```
